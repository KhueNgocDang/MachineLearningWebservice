```python
import json
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from datetime import datetime
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from datetime import datetime
import joblib
```

```python
df=pd.read_csv('Stockprice.csv')
##df.drop(['<OpenFixed>','<HighFixed>','<LowFixed>','<CloseFixed>','<VolumeDeal>','<VolumeFB>','<VolumeFS>'],1,inplace = True)
##df[['date']] = df[['<DTYYYYMMDD>']].applymap(str).applymap(lambda s: "{}/{}/{}".format(s[4:6],s[6:], s[0:4]))
df.drop(['DTYYYYMMDD'],1, inplace = True)
df.drop(['Year'],1,inplace = True)
df.drop(['Doy'],1,inplace=True)
df
```

```python
##df['month'] = df['date'].dt.month

df['2 Days STD']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=2).std().shift(-1))
df['2 Days MA']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=2).mean().shift(-1))
df['H-L']=df['High']-df['Low']
df['C-O']=df['Close']-df['Open']
df['7 Days STD']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=7).std().shift(-6))
df['7 Days MA']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=7).mean().shift(-6))
df['14 Days MA']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=14).mean().shift(-13))
df['21 Days MA']=df.groupby('Ticker').Close.apply(lambda x: x.rolling(window=21).mean().shift(-20))
df.replace([np.inf,-np.inf],np.nan,inplace=True)
df.dropna(inplace=True)
df.tail(22)
```

```python
X = df[['Volume','21 Days MA','14 Days MA','7 Days MA','7 Days STD']]
Y = df[['Close']]
X
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = .20)
```

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred=lr_model.predict(x_test)
```

```python
kfold = model_selection.KFold(n_splits=20, random_state =100)
result_kfold = model_selection.cross_val_score(lr_model,x_test,y_test,cv=kfold)
print("Accuracy: ", result_kfold.mean()*100)
```
